So, values below the lower limit and values beyond the upper limit are considered as outliers, and we need to either trim the outlier values or we need to replace outliers below the lower limit with the lower limit value and values beyond the upper limit with the upper limit value, respectively.

```python
# To find the number of outliers with upper limit and lower limit in each column, respectively.
col = ['actual_time_cumulative', 'osrm_time_cumulative', 'actual_distance_to_destination_cumulative', 'segment_actual_time'
                 , 'osrm_distance_cumulative','segment_osrm_time','segment_osrm_distance','Actual_Average_Speed','Total_trip_distance',
                 'Actual_trip_time','Osrm_Average_Speed','Total_osrm_trip_distance','No_of_stops','Trip_total_duration(mins)']
def find_outlier(data, column):
    
    percentile25 = data[column].quantile(.25)
    percentile75 = data[column].quantile(.75)
    IQR = percentile75 - percentile25
    upperlimit = percentile75 + 1.5*IQR
    lowerlimit = percentile25 - 1.5*IQR 
    # outlier = data[(data[column] > upperlimit) | (data[column] < lowerlimit)]
    outlier_lower = data[(data[column] < lowerlimit)]
    outlier_upper = data[(data[column] > upperlimit)]

    return outlier_lower,outlier_upper, upperlimit, lowerlimit

for column in col:
    outlier_lower,outlier_upper, upperlimit, lowerlimit = find_outlier(df,column)
    print(f"Upper Limit of '{column}' = {upperlimit}")
    print(f"Lower Limit of '{column}' = {lowerlimit}")
    print(f"Column '{column}' has = {len(outlier_lower)} lower outliers.")
    print(f"Column '{column}' has = {len(outlier_upper)} upper outliers.")
    print('-----------------------------------------------------')
```

```python
for column in col:
    outlier_lower,outlier_upper, upper_limit, lower_limit = find_outlier(df, column)

    # print(f"[BEFORE] Column '{column}' has {len(outlier_lower)} lower and {len(outlier_upper)} upper outliers."
    
    # Cap values above the upper limit
    df[column] = np.where(df[column] > upper_limit, upper_limit, df[column])
    
    # Drop rows with values below the lower limit
    df = df[df[column] >= lower_limit]

    remaining_upper = df[df[column] > upper_limit]
    remaining_lower = df[df[column] < lower_limit]

    print(f"Column '{column}' has {len(remaining_lower)} lower and {len(remaining_upper)} upper outliers.")
    print('-' * 60)
```
